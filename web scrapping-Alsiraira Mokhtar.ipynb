# Task 1

```python
from bs4 import BeautifulSoup
import requests
import sys
```

https://www.ibm.com/topics/machine-learning

```python
# Set the user-agent
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"

headers = {"user-agent": USER_AGENT}
r = requests.get("https://www.ibm.com/topics/machine-learning", headers=headers)

print("Status code: ", r.status_code)

if r.status_code == 200:
    soup = BeautifulSoup(r.content, "html.parser")
else:
    print("Code is not valid")
    sys.exit(0)

# Extract headline 
headline = soup.title

# Extract text content 
text_content = ""
for paragraph in soup.find_all("p"):
    text_content += paragraph.text + "\n"

# Print the extracted headline and text content
print("Headline: ", headline.prettify())
print("Text Content: ", text_content)
```

```python
# Save the data to a file
with open("output1.txt", "w", encoding="utf-8") as file:
    file.write("Headline: " + str(headline) + "\n")
    file.write("Text Content: " + text_content)
    print("Data saved to output1.txt")
```

```python
# List of city names and their URLs
cities = {
    "London": "https://www.timeanddate.com/worldclock/uk/london",
    "New York": "https://www.timeanddate.com/worldclock/usa/new-york",
    "Tokyo": "https://www.timeanddate.com/worldclock/japan/tokyo",
    "Sydney": "https://www.timeanddate.com/worldclock/australia/sydney",
    "Paris": "https://www.timeanddate.com/worldclock/france/paris",
    "Dubai": "https://www.timeanddate.com/worldclock/united-arab-emirates/dubai",
    "Beijing": "https://www.timeanddate.com/worldclock/china/beijing",
    "Rio de Janeiro": "https://www.timeanddate.com/worldclock/brazil/rio-de-janeiro",
    "Cape Town": "https://www.timeanddate.com/worldclock/south-africa/cape-town",
    "Toronto": "https://www.timeanddate.com/worldclock/canada/toronto",
    "Khartoum": "https://www.timeanddate.com/worldclock/sudan/khartoum"
}

# Set the user-agent 
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"
headers = {"user-agent": USER_AGENT}

# Open a file to save the data
with open("weather_info_timeanddate.txt", "w") as file:
    # Iterate over cities and extract weather information
    for city, url in cities.items():
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  # Raise Error for bad requests
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract weather details using class and ID names
            current_temperature = soup.find('div', id='wt-tp').text.strip()

            # Write the city name and temperature to the file
            file.write(f"Weather Information for {city}:\n")
            file.write(f"Current Temperature in celsius: {current_temperature}\n\n")

            # Print the extracted information 
            print(f"Weather Information for {city}:")
            print(f"Current Temperature in celsius: {current_temperature}\n")

        except requests.exceptions.HTTPError as errh:
            print(f"HTTP Error occurred: {errh}")
        except requests.exceptions.ConnectionError as errc:
            print(f"Error Connecting: {errc}")
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error: {errt}")
        except requests.exceptions.RequestException as err:
            print(f"An error occurred: {err}")
        except AttributeError as attr_err:
            print(f"Attribute Error occurred: {attr_err}")

print("Weather information from |www.timeanddate.com| saved to weather_info_timeanddate.txt")
```

```python
# Set the user-agent
USER_AGENT = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36"
headers = {"User-Agent": USER_AGENT}

# List of cities and their URLs
cities = {
    "London": "https://www.wunderground.com/weather/gb/london",
    "New York": "https://www.wunderground.com/weather/us/ny/new-york",
    "Tokyo": "https://www.wunderground.com/weather/jp/tokyo",
    "Sydney": "https://www.wunderground.com/weather/au/nsw/sydney",
    "Paris": "https://www.wunderground.com/weather/fr/ile-de-france/paris",
    "Dubai": "https://www.wunderground.com/weather/ae/dubai",
    "Beijing": "https://www.wunderground.com/weather/cn/beijing",
    "Rio de Janeiro": "https://www.wunderground.com/weather/br/rj/rio-de-janeiro",
    "Cape Town": "https://www.wunderground.com/weather/za/western-cape/cape-town",
    "Toronto": "https://www.wunderground.com/weather/ca/on/toronto",
    "Moscow": "https://www.wunderground.com/weather/ru/moscow"
}

# Open a file and save the data
with open("weather_info_wunderground.txt", "w") as file:
    # Iterate over cities and fetch weather information
    for city, url in cities.items():
        try:
            response = requests.get(url, headers=headers)
            response.raise_for_status()  # Raise Error for bad requests
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract temperature details
            temperature_element = soup.find('div', class_='current-temp').find('span', class_='wu-value-to')
            current_temperature_fahrenheit = int(temperature_element.text.strip())

            # Write city name and temperature to the file
            file.write(f"Weather Information for {city}:\n")
            file.write(f"Current Temperature in fahrenheit: {current_temperature_fahrenheit}°F\n\n")

            # Print extracted information (optional)
            print(f"Weather Information for {city}:")
            print(f"Current Temperature in fahrenheit: {current_temperature_fahrenheit}°F\n")

        except requests.exceptions.HTTPError as errh:
            print(f"HTTP Error occurred: {errh}")
        except requests.exceptions.ConnectionError as errc:
            print(f"Error Connecting: {errc}")
        except requests.exceptions.Timeout as errt:
            print(f"Timeout Error: {errt}")
        except requests.exceptions.RequestException as err:
            print(f"An error occurred: {err}")
        except AttributeError as attr_err:
            print(f"Attribute Error occurred: {attr_err}")

print("Weather information from |www.wunderground.com| saved to weather_info_wunderground.txt")
```
